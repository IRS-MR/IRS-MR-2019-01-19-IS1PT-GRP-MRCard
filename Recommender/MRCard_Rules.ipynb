```python
for index, row in df.iterrows():
    cardid = row['credit_card_id']
    age__min = row['age_min']
    age__max = row['age_max']
    gender = row['gender_req']
    annual_income_singaporean_min = row['annual_income_singaporean_min']
    annual_income_pr_min = row['annual_income_pr_min']
    annual_income_foreigner_min = row['annual_income_foreigner_min']
    overall_cashback_min_spend = row['overall_cashback_min_spend']
    
    class Person(Fact):
        age = Field(int)
        gender = Field(str)
        citizenship = Field(str)
        annual_income = Field(int)
        total_spending_amount = Field(int)
        pass
    
    class Eligibility(KnowledgeEngine):
        
        @Rule(AND(
            AS.v << Person(data__age = MATCH.data__age),
            TEST(lambda data__age: data__age >= age__min),
            AS.v << Person(data__age = MATCH.data__age),
            TEST(lambda data__age: data__age <= age__max),
            AS.v << Person(data__total_spending_amount = MATCH.data__total_spending_amount),
            TEST(lambda data__total_spending_amount: data__total_spending_amount >= overall_cashback_min_spend),
            OR(
                AND(
                AS.v << Person(data__citizenship = MATCH.data__citizenship),
                TEST(lambda data__citizenship: data__citizenship == 'Singaporean'),
                AS.v << Person(data__annual_income = MATCH.data__annual_income),
                TEST(lambda data__annual_income: data__annual_income >= annual_income_singaporean_min),    
                ),
                AND(
                AS.v << Person(data__citizenship = MATCH.data__citizenship),
                TEST(lambda data__citizenship: data__citizenship == 'PR'),
                AS.v << Person(data__annual_income = MATCH.data__annual_income),
                TEST(lambda data__annual_income: data__annual_income >= annual_income_pr_min),    
                ),
                AND(
                AS.v << Person(data__citizenship = MATCH.data__citizenship),
                TEST(lambda data__citizenship: data__citizenship == 'Foreigner'),
                AS.v << Person(data__annual_income = MATCH.data__annual_income),
                TEST(lambda data__annual_income: data__annual_income >= annual_income_foreigner_min),    
                )
            )
            )
         )
    def age_check(self, v):
        print(cardid + ": available")
        df_result = df_result.append({'cardid':cardid, 'comment':"available"}, ignore_index=True)
```
